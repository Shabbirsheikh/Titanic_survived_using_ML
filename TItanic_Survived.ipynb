```python
import numpy as np
import pandas as pd
```

```python
df1=pd.read_csv('train.csv') #reading train.csv file and making dataframe
```

```python
df2=pd.read_csv('test.csv') #reading test.csv file and making dataframe
```

```python
df1.head()
```

```python
inputs1_1= df1['PassengerId']
inputs1_2 = df1['Sex' ]
inputs1_3 = df1['Age']
inputs1_4 = df1['Fare']
target1 = df1['Survived']
#Taking only necessary columns
```

```python
df2.head()
```

```python
inputs2_1= df2['PassengerId']
inputs2_2 = df2['Sex' ]
inputs2_3 = df2['Age']
inputs2_4 = df2['Fare']
```

```python
inputs1 = pd.concat([inputs1_1,inputs1_2,inputs1_3,inputs1_4],axis = 1)
#concating it to new dataframe
```

```python
inputs2 = pd.concat([inputs2_1,inputs2_2,inputs2_3,inputs2_4],axis = 1)
```

```python
inputs1
```

```python
target1
```

```python
inputs2.head()
```

```python
import math
median_Age1=math.floor(inputs1.Age.median())
median_Age2=math.floor(inputs2.Age.median())
median_Fare1=math.floor(inputs1.Fare.median())
median_Fare2=math.floor(inputs2.Fare.median())
# taking median of the coulmn having NAN values
```

```python
inputs1.Age = inputs1.Age.fillna(median_Age1)
inputs2.Age = inputs2.Age.fillna(median_Age2)
inputs1.Fare = inputs1.Fare.fillna(median_Fare1)
inputs2.Fare = inputs2.Fare.fillna(median_Fare2)
```

```python
inputs1.Age = inputs1.Age.fillna(median_Age1)
#Rplacing NAN values to the median of that column 
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
inputs1['Sex_n'] = le_Sex.fit_transform(inputs1['Sex'])
inputs2['Sex_n'] = le_Sex.fit_transform(inputs2['Sex'])
#As the sex column is not having new merical values so using label encoder we will 0 and 1 to Male and female or vicversa
```

```python
inputs_n1 = inputs1.drop('Sex',axis = 1)
inputs_n2 = inputs2.drop('Sex',axis = 1)
inputs_n1
```

```python
inputs_n2
```

```python
from sklearn.model_selection import train_test_split
X_train = inputs_n1.drop(['PassengerId'],axis=1)
Y_train = target1
X_test = inputs_n2.drop(['PassengerId'],axis=1)
#Now training our model
```

```python
X_test.isnull().sum()
```

```python
X_train.isnull().sum()
#As we do not having any null values we can do further classification model
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
#I am using Decision tree classification we can use any other classification model but I generally prefer Decision tree or SVM 
```

```python
model.fit(X_train,Y_train)
#Fitting our model
```

```python
Y_pred = model.predict(X_test)
#predicting Y_predict from X_test
```

```python
acc_model = round(model.score(X_train, Y_train) * 100, 2)
acc_model 
#Checking our score
```
